**Source**: Kaggle - Spotify Data: Popular Hip-hop Artists and Tracks

**Column Descriptors:**
- Artist: The name of the artist, offering a direct link to the creative mind behind the track.
- Track Name: The title of the track, encapsulating its identity and essence.
- Popularity: A numeric score reflecting the track's reception and appeal among Spotify listeners.
- Duration (ms): The track's length in milliseconds, detailing the temporal extent of the musical experience.
- Track ID: A unique identifier within Spotify's ecosystem, facilitating direct access to the track for further exploration.


```python
import pandas as pd
data = pd.read_csv("top_hiphop_artists_tracks.csv")
data
```

```python
duplicates = data[data.duplicated()]
duplicates
```

###### When  I  run the duplicated code  for the first time, I didn't get accurate results. The code seemed to include songs diffrent songs by the same artist as duplicates. I chose to identify duplicates by using the track name to get more accurate results.

```python
# Find duplicates based on Track Name
duplicates = data[data.duplicated(subset=['Track Name'], keep=False)]
duplicates
```

```python
# Drop duplicates based on Track Name, keeping the first occurrence
data_cleaned = data.drop_duplicates(subset=['Track Name'], keep='first')
data_cleaned
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Top 10 most popular tracks
top_tracks = data_cleaned.sort_values(by='Popularity', ascending=False).head(10)
top_tracks
```

```python
# Top 10 most popular artists
top_artists = data_cleaned.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(10)

top_artists
```

```python
from tabulate import tabulate

# Sort the data by Popularity and get the top 10 tracks
top_tracks_overall = data.nlargest(10, 'Popularity')

# Display the top 10 tracks overall in a table format
table = top_tracks_overall[['Artist', 'Track Name', 'Popularity']].reset_index(drop=True)
print(tabulate(table, headers='keys', tablefmt='github'))
```

```python
# Sort the data by Popularity and get the top 10 tracks
top_tracks_overall = data.nlargest(10, 'Popularity')

# Display the top 10 tracks overall
print("Top 10 Most Popular Tracks Overall:")
for index, row in top_tracks_overall.iterrows():
    print(f"{row['Artist']} - {row['Track Name']} (Popularity: {row['Popularity']})")

```

```python
# Scatter plot of Popularity vs. Track Duration
plt.figure(figsize=(12, 6))
plt.scatter(data['Duration (ms)'], data['Popularity'], alpha=0.5)
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.title('Popularity vs. Track Duration')
plt.grid(True)
plt.show()
```

## Observation:

#### There is a significant concentration of tracks in the dataset with durations between 100,000 ms and 300,000 ms.

## Interpretation:

#### This concentration suggests that a large proportion of the tracks in the dataset fall within a specific duration range, indicating a common trend or preference in track duration.
